```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#importing November Youtube raw data
Nov_df = pd.read_csv('../Resources/NOV_YOUTUBE_INFLUENCER_RAW.csv')
Nov_df.head()
```

```python
#looking for duplicates
duplicate_values = Nov_df['Youtube channel'].duplicated()
print(duplicate_values)
```

```python
#droping duplicates
df = Nov_df.drop_duplicates(subset=['Youtube channel'], keep='first')
df
```

```python
#updating column names
renamed_df = df.rename(columns={"Youtube channel":"Youtube_channel", "youtuber name":"Youtuber_name",
                                    "Followers":"Subscribers","\nViews (Avg.)":"Average_view",
                                    "Likes (Avg.)":"Average_likes","Comments (Avg.)":"Average_comments",
                                    "Category-2":"Sub-category"})
renamed_df.head()
```

```python
#replacing empty cells with "other" for category and "unknown" for country  
cleaned_Nov_df= renamed_df.fillna({"Country": 'Unknown',
                                       "Category": 'Other',
                                      "Sub-category":'Other'})
cleaned_Nov_df
```

```python

filtered_df = cleaned_Nov_df[cleaned_Nov_df['Country'] != 'Unknown']
filtered_df
```

```python
#dropping the number column
cleaned_Nov_df.drop(columns=['s.no'])
```

```python
# to save file as csv to Analysis file
#cleaned_Nov_df.to_csv("../Analysis/Nov_Youtube.csv", encoding='utf8', index=False)
cleaned_Nov_df.to_csv("../Resources/Nov_clean_data.csv", index=False, header=True)
```
