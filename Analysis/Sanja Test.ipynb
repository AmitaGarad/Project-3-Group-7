```python
#importing December Youtube data
df_csv = pd.read_csv('../Resources/DEC_YOUTUBE_INFLUENCER_RAW.csv')
df_csv
```

```python
#looking for dupliates
duplicate_values = df_csv['Youtube channel'].duplicated()
print(duplicate_values)
```

```python
#dropping duplicates
Dec_df=df_csv.drop_duplicates(subset=['Youtube channel'], keep='first')
Dec_df
```

```python
#updating column names
renamed_df = Dec_df.rename(columns={"Youtube channel":"Youtube_channel", "youtuber name":"Youtuber_name",
                                    "Followers":"Subscribers","Views (Avg.)":"Average_view",
                                    "Likes (Avg.)":"Average_likes","Comments (Avg.)":"Average_comments",
                                    "Category-2":"Sub-category"})
renamed_df.head()
```

```python
#replacing blank values
cleaned_Dec = renamed_df.fillna({"Category": 'Others',
                                      "Country": 'Unknown',
                                      "Sub-category": 'Others'})
cleaned_Dec.head()
```

```python
#dropping the number column
cleaned_Dec.drop(columns=['s.no'])
```
