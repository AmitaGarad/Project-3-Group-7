```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('../Resources/AI_Influencer_cleaned.csv').head(200)
df.to_json("../Resources/output/AI_Influencer_200.json")
df
```

```python
df2 = pd.read_csv('../Resources/Sept_clean_data.csv')
df2 = df2.rename(columns={'S.no':'s.no'})
df2
```

```python
df3 = pd.read_csv('../Resources/Nov_clean_data.csv')
df3 = df3.rename(columns={'\nViews (Avg.)':'Average_view'})
df3
```

```python
df4 = pd.read_csv('../Resources/Dec_clean_data.csv')
df4
```

```python
#Convert views and likes from string to value
def valueChange(x):
    if 'K' in x:
        return float(x.replace('K', '')) * 1000
    elif 'M' in x:
        return float(x.replace('M', '')) * 1000000
    elif 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    else:
        return float(x)
```

```python
df_combine = pd.concat([df2,df3,df4], ignore_index=True)
```

```python
df_combine['Average_view'] = df_combine['Average_view'].apply(valueChange)
df_combine['Average_likes'] = df_combine['Average_likes'].apply(valueChange)
df_combine = df_combine[~df_combine['Category'].isin(['Other', 'Others'])]
df_combine
```

```python
df_combine.dtypes
```

```python
df_CAT = df_combine.groupby(['Category'])
grouped_df = df_CAT[['Average_view','Average_likes']].mean().reset_index()
grouped_df 
```

```python
grouped_df['Average_view'] = grouped_df['Average_view']/grouped_df['Average_view'].max()
grouped_df['Average_likes'] = grouped_df['Average_likes']/grouped_df['Average_likes'].max()
grouped_df
```

```python
grouped_df = grouped_df[~grouped_df['Category'].isin(['Other', 'Others'])]
grouped_df
```

```python
grouped_df.to_csv("../Resources/output/Category_Summary.csv", index=False, header=True)
grouped_df.to_json("../Resources/output/Category_Summary.json")
```
