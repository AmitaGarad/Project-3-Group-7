```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#importing September Youtube raw data
Sept_df = pd.read_csv('../Resources/SEP_YOUTUBE_INFLUENCER_RAW.csv')
Sept_df
```

```python
#looking for duplicates
duplicate_values = Sept_df['Name'].duplicated()
print(duplicate_values)
```

```python
#removing duplicates
df = Sept_df.drop_duplicates(subset=['Name'], keep='first')
df
```

```python
#updating column names
renamed_Sept_df = df.rename(columns={"Name":"Youtube_channel", "Youtuber":"Youtuber_name",
                                     "Avg. views":"Average_view","Avg. likes":"Average_likes",
                                     "Avg Comments":"Average_comments","Category_2":"Category",
                                    "Category_3":"Sub-category", "Unnamed: 10":"Date"})
renamed_Sept_df.head()
```

```python
#rearanging the columns to match the file in the Clean_Data
renamed_Sept_df=renamed_Sept_df[["S.no","Youtube_channel","Youtuber_name","Category",'Subscribers',
                                 "Country","Average_view","Average_likes","Average_comments",
                                 "Sub-category","Date"]]
renamed_Sept_df
```

```python
#replacing empty cells with "other" for category and "unknown" for country  
cleaned_Sept_df= renamed_Sept_df.fillna({"Country": 'Unknown',
                                       "Category": 'Other',
                                      "Sub-category":'Other'})
cleaned_Sept_df
```

```python
#dropping the number column
cleaned_Sept_df.drop(columns=['S.no'])
```
